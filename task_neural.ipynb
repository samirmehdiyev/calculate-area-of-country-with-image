```python
import numpy as np
import random
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('aze.png')

# BGR -> RGB
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.imwrite('map.png', img)
plt.imshow(img)
```

```python
INTERVAL = 100000

inside = 0
outside = 0

points_array = []
HEIGHT, WIDTH, CHANNELS = image.shape
REAL_HEIGHT = 388.806  # kms
RATIO = REAL_HEIGHT / HEIGHT
REAL_WIDTH = WIDTH * RATIO
REAL_AREA = REAL_HEIGHT * REAL_WIDTH
for i in range(INTERVAL):

    rand_x = random.randint(0, WIDTH - 1)
    rand_y = random.randint(0, HEIGHT - 1)
#     print(img)
    if (img[rand_y, rand_x] == np.array([0, 0, 0])).all():
        outside += 1
    else:
        inside += 1
    points_array.append((rand_x, rand_y))

plt.imshow(img)
plt.scatter(*zip(*points_array), marker='.', color="red")
plt.show()


print(inside)
```

```python
area = (inside/INTERVAL)*REAL_AREA
print(area)
```
